```python
import pandas as pd
import os
from IPython.core.display import display, HTML

def multi_column_df_display(list_dfs, cols=3):
    html_table = "<table style='width:100%; border:0px'>{content}</table>"
    html_row = "<tr style='border:0px'>{content}</tr>"
    html_cell = "<td style='width:{width}%;vertical-align:top;border:0px'>{{content}}</td>"
    html_cell = html_cell.format(width=100/cols)

    cells = [ html_cell.format(content=df.to_html()) for df in list_dfs ]
    cells += (cols - (len(list_dfs)%cols)) * [html_cell.format(content="")] # pad
    rows = [ html_row.format(content="".join(cells[i:i+cols])) for i in range(0,len(cells),cols)]
    display(HTML(html_table.format(content="".join(rows))))

csv_path = '/home/thiagodepaulo/exp/PyPBG/exp/results/CSTR/'
l = []
for filename in os.listdir(csv_path):    
    if filename.endswith('.csv'):        
        df = pd.read_csv(csv_path+filename, index_col=0, header=0)
        s = filename.replace('.csv','').split('_')
        df['num_samples'] = int(s[1])
        df['num_exp'] = int(s[2])
        df['cls_name'] = df.index[1]
        l.append(df)
df = pd.concat(l)
```

```python
df
```

```python
def f(x):
    return pd.Series(dict(precision = x['precision'].max(), 
                        recall = x['recall'].max(),   
                        f1_score = x['f1-score'].max(),                      
                        cls_name = set(x['cls_name'])))
df_acc = df.loc[['accuracy']].groupby(['num_samples', 'num_exp']).apply(f)
df_macro = df.loc[['macro']].groupby(['num_samples', 'num_exp']).apply(f)
df_w = df.loc[['weighted']].groupby(['num_samples', 'num_exp']).apply(f)
```

```python
print("\t\t\t\t\tAs tabelas são respectivamente Accuracy, Macro e Weighted")
multi_column_df_display([df_acc, df_macro, df_w], cols=3)
```
